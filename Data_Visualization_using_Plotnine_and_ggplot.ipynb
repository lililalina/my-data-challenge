Import library

```
import seaborn as sns
import pandas as pd
from pandas.plotting import scatter_matrix
from matplotlib import pyplot as pyplot
from matplotlib import font_manager as font_manager
from pandas.plotting import  autocorrelation_plot
import plotnine
from plotnine import ggplot, geom_line, aes

import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go

import numpy as np
from datetime import datetime, timedelta
```

Line Plot

```
work_years = [1,2,3,4,5]
income = [6500, 7500, 9000, 10000, 12000]

df = pd.DataFrame({'Years at work': work_years, 'income': income})
```

```
df.head()
```

```
line_plot = ggplot(data=df, mapping=aes(x="Years at work", y="income"))
```

```
(line_plot + plotnine.geom_line(color = "blue") + plotnine.labs(title= "Simple line plot(Years/income)"))
```

#Multiline Plot

```
work_years = [1,2,3,4,5]
all_engineers_income = [6500, 7500, 9000, 10000, 12000]
sde_income = [5500, 6500, 7500, 8500, 10000]

income_df = pd.DataFrame({'Years at work': work_years, 'All Engineers Income': all_engineers_income, 'SDE Income': sde_income})
```

```
income_df.head()
```

```
(ggplot()+
 plotnine.geom_line(data = income_df,mapping=aes(x="Years at work", y = "All Engineers Income"), color = "blue")+
 plotnine.geom_line(data = income_df,mapping=aes(x="Years at work", y = "SDE Income"), color = "green")+
 plotnine.labs(title="Multiline plot from ML and SDE Engineers Income/years")+
 plotnine.scale_alpha_identity(guide="legend", name="Legend", breaks=["blue", "green"], labels=['ML Engineers Indome', 'SDE Income']))
```

# Bar Plot

```
df = sns.load_dataset('titanic')
df.head()
```

```
(plotnine.ggplot(data=df.dropna(), mapping=aes(x="pclass", y="fare", fill='sex'))+
plotnine.geom_col()+
plotnine.labs(title="Bar Plot of Passenger class vs fare"))
```

# Scatter Plot

```
#scatter plot
scatter_plot = ggplot(data=df, mapping=aes(x="pclass", y="fare"))

(scatter_plot +
 plotnine.labs(title="Scatter Plot of Passenger class vs fare") +
 plotnine.geom_point(shape='o', size = 7, color = 'green', alpha = 0.7))
```

# Histogram


```
hist = ggplot(data=df, mapping = aes(x='age'))

(hist +
 plotnine.labs(title = "Histogram of Passenger Age")+
 plotnine.geom_histogram()+
 plotnine.geom_rug())
```

# Density Plot

```
kde = ggplot(data=df, mapping = aes(x='age'))

(kde +
 plotnine.labs(title = "Density Plot of Passenger Age")+
 plotnine.geom_density())
```

# Box Plot


```
box_plot = ggplot(data=df, mapping = aes(x='embarked', y='fare'))

(box_plot +
 plotnine.labs(title = "Box Plot for Embarked and Fare")+
 plotnine.geom_boxplot())
```

# Violin Plot

```
violinplot = ggplot(data=df, mapping = aes(x='sex', y='age'))

(violinplot +
 plotnine.geom_violin()+
 plotnine.labs(title = "Violin Plot for Passenger Age by Gender")+
 plotnine.geom_jitter(position = plotnine.position_jitter(0.3)))
```
